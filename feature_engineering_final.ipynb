# Post modelling feature analysis

```python
train=pd.read_csv('../data/final_train.csv')
test=pd.read_csv('../data/final_test.csv')
```

```python
train
```

```python
img_cols = [f"dino_{i}" for i in range(768)]
sat_cols= [f"sat_feat_{i}" for i in range(256)]
target = 'price' # Ensure this matches your target column name
tab_cols = [c for c in train.columns if c not in img_cols + [target]+sat_cols]
```

```python
train.drop(sat_cols,axis=1,inplace=True)
test.drop(sat_cols,axis=1,inplace=True)
```

```python
train
```

```python
from sklearn.ensemble import IsolationForest
```

```python
# Isolation Forest for outlier detection
iso_forest = IsolationForest(contamination=0.01, random_state=42)

# Select numerical features for outlier detection (excluding price and categorical)
features_for_outlier = train.select_dtypes(include=['int64', 'float64']).drop(columns=['price','waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated','sqft_living15',
       'sqft_lot15', 'year', 'month', 'time_since_built',
       'time_since_renovated','month_sin','month_cos']+img_cols)

# Fit and predict
outlier_labels = iso_forest.fit_predict(features_for_outlier)
```

```python

# -1 for outliers, 1 for inliers
print(f"Number of outliers detected: {(outlier_labels == -1).sum()}")
print(f"Number of inliers: {(outlier_labels == 1).sum()}")
```

```python
# remove them and make a new dataframe final_train_no_outliers
final_train_no_outliers = train[outlier_labels == 1].reset_index(drop=True)
```

```python
# save to csv
final_train_no_outliers.to_csv('../data/final_train_no_outliers.csv',index=False)
```

# Outlier Removal

```python
final_train_no_outliers
```

```python
# add two feature lat+lon and lat-lon as lat45 and lon45
final_train_no_outliers['lat_plus_lon'] = final_train_no_outliers['lat'] + final_train_no_outliers['long']
final_train_no_outliers['lat_minus_lon'] = final_train_no_outliers['lat'] - final_train_no_outliers['long']
```

```python
test['lat_plus_lon'] = test['lat'] + test['long']
test['lat_minus_lon'] = test['lat'] - test['long']
```

# Geoclusters

```python
from sklearn.cluster import KMeans
```

```python
# 1. Initialize the model
# For 16k rows, 100 clusters is a good starting point to capture local pockets.
kmeans = KMeans(n_clusters=100, random_state=42, n_init=10)
```

```python
# 2. Fit ONLY on training data
# Assuming your coordinates are 'latitude' and 'longitude'
train_coords = final_train_no_outliers[['lat', 'long']]
kmeans.fit(train_coords)
```

```python
final_train_no_outliers['geo_cluster'] = kmeans.predict(train_coords)
```
